```python
import time
import torch
import numpy as np
from torch.nn import Conv1d
from tqdm import tqdm
```

```python
time_lst = []
m = Conv1d(4, 4, 2)

for i in tqdm(np.arange(2, 1000)):
    elpased_time = 0
    for j in range(100):
        time_start = time.time()
        dummy_inpput = torch.randn(128, 4, i)
        m(dummy_inpput)
        time_end = time.time()
        elpased_time += time_end - time_start
    time_lst.append(elpased_time / 100)

print(time_lst)
```

```python
from matplotlib import pyplot as plt

plt.plot(np.arange(2, 1000), time_lst)
plt.xlabel("Trajectory Length")
plt.ylabel("Inference Time")
```
